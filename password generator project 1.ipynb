```python
from tkinter import*
from functools import partial
import random
screen=Tk()
my_image=PhotoImage(file="padlock.png")
img_lbl=Label(screen,image=my_image,bg="yellow",borderwidth=0)
img_lbl.place(x=320,y=70)
screen.geometry("450x450")
screen.maxsize(width=800,height=620)
screen.minsize(width=800,height=620)
screen.config(background="yellow")
screen.title("PASSWORD GENERATOR")
heading=Label(screen,text="STRONG PASSWORD GENERATOR",font=("Algerian 15 bold")).place(x=60,y=15)
scroll=Scrollbar(screen)
scroll.pack(side=RIGHT,fill=Y)
my_menu=Menu(screen)
screen.config(menu=my_menu)
file_menu=Menu(screen)
my_menu.add_cascade(label="file",menu=file_menu)

def generate(upr,lwr,nbr,spchar,val):
    
    up_case="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lw_case="abcdefghijklmnopqrstuvwxyz"   
    num="0123456789" 
    sp_char="@#$%^&*"
    all=""
    if upr.get()==1:
        all=all+up_case
    if lwr.get()==1:
        all=all+lw_case   
    if nbr.get()==1:
        all=all+num
    if spchar.get()==1:
        all=all+sp_char
    password="".join(random.sample(all,val.get()))
    pw.delete(0,END)         #this line is for refreshing existing length of password
    pw.insert(0,password)
    return
    
    
upr=IntVar()
lwr=IntVar()
nbr=IntVar()
spchar=IntVar()
val=IntVar()
l1=Label(screen,text="Enter a length of password",bd=5).place(x=20,y=60)
l2=Label(screen,text="STRONG PASSWORD IS HERE:-",bd=8,bg="green",fg="white").place(x=20,y=370)
c1=Checkbutton(screen,text="characters in uppercase  eg-ABCD ",variable=upr,onvalue=1,offvalue=0).place(x=180,y=120)
c2=Checkbutton(screen,text="characters in Lowercase  eg-abcd ",variable=lwr,onvalue=1,offvalue=0).place(x=180,y=160)
c3=Checkbutton(screen,text="characters in Numbers    eg-0123 ",variable=nbr,onvalue=1,offvalue=0).place(x=180,y=200)
c3=Checkbutton(screen,text="special characters       eg-@#$% ",variable=spchar,onvalue=1,offvalue=0).place(x=180,y=240)
s1=Spinbox(screen,from_=8,to_=24,textvariable=val,bd=6)
s1.place(x=180,y=60)
global val1
val1=IntVar
pw=Entry(screen,textvariable=val1,bd=8,bg="green",fg='white')
pw.place(x=220,y=370)
generate=partial(generate,upr,lwr,nbr,spchar,val)
B1=Button(screen,text="GENERATE PASSWORD",bd=9,bg="blue",fg="white",command=generate)
B1.place(x=210,y=300)
screen.mainloop()
```
